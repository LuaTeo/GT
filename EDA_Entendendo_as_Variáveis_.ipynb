Agora, entendemos nosso conjunto de dados em geral e verificamos se há valores faltantes. A próxima parte dessa jornada é a visualização de dados! Nosso objetivo é realizar análises para verificar a distribuição e relação entre as variáveis.

Utilizaremos a biblioteca seaborn para visualização de dados estatísticos. Seaborn é uma biblioteca de visualização de dados baseada em matplotlib e minha favorita com facilidade de uso.

###Quantas casas tem vista pro mar ou não?

```python
df.groupby(['waterfront']).count()
```

```python
#Vista pro mar vs total
sns.countplot(x='waterfront', data=df)
```

###Há mais casas com vista boa ou ruim?

```python
sns.countplot(x='view', data=df)
```

### E se fizermos a distinção da condição das casas considerando se tem vista pro mar ou não?

```python
df.groupby(['view','waterfront']).count()
```

```python
sns.countplot(x='view', hue='waterfront', data=df)
```

###Como os dados da quantidade de quartos se distribuem? E o tamanho da área interna?

```python
df['bedrooms'].describe()
```

```python
sns.displot(df['bedrooms'], kde = True);
```

```python
df['sqft_living'].describe()
```

```python
sns.displot(df['sqft_living'], kde = True);
```

###Qual casa é mais cara? Com vista pro mar ou não?

```python
df.groupby('waterfront')['price'].describe()
```

```python
#Valor Total
sns.barplot(x="waterfront", y="price", data=df, estimator=np.sum)
```

```python
#Valor Médio
sns.barplot(x="waterfront", y="price", data=df, estimator=np.mean)
```

###Qual a quantidade de banheiros que faz a casa ser mais cara?

```python
#Valor Total
sns.barplot(x="bedrooms", y="price", data=df, estimator=np.sum)
```

```python
#Valor Médio
sns.barplot(x="bedrooms", y="price", data=df, estimator=np.mean)
```

###Vamos verificar algumas correlações...

```python
corr = df.corr(method='pearson')
corr
```

```python
df_filter = df[['price', 'bedrooms', 'bathrooms','view', 'waterfront', 'sqft_living']]
```

```python
corr = df_filter.corr(method='pearson')

# heatmap - gráfico de calor
plt.figure(figsize=(8,4))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm')
plt.show()
```

```python
sns.pairplot(df_filter, hue='price')
```

```python
# relação entre a variável x e y
plt.figure(figsize=(12, 7))
sns.scatterplot(x="bedrooms", y="price", hue="waterfront", data=df_filter)
```
